```python
import csv
from tqdm import tqdm
from uszipcode import SearchEngine
import json
import matplotlib.pyplot as plt
import requests
import numpy as np
import math

zipsearch = SearchEngine(simple_zipcode=True)
```

```python
response = requests.get("https://data.medicare.gov/api/views/xubh-q36u/rows.csv?accessType=DOWNLOAD")
with open("HospitalGeneralInfo.csv", 'wb') as f:
    f.write(response.content)
response = requests.get("https://simplemaps.com/static/data/us-cities/uscitiesv1.4.csv")
with open("CitiesInfo.csv", 'wb') as f:
    f.write(response.content)
```

```python
hospitals = []
with open("HospitalGeneralInfo.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in tqdm(reader):
        h = {}
        h["state"] = row["State"]
        if h["state"] not in ["HI", "AK", "PR"]:
            zipcode = zipsearch.by_zipcode(row["ZIP Code"])
            zj = json.loads(zipcode.to_json())
            if zj["lat"]:
                h["lat"] = float(zj["lat"])
                h["lng"] = float(zj["lng"])
                h["name"] = row["Hospital Name"]
                h["city"] = row["City"]
                h["score"] = row["Hospital overall rating"]
                if h["score"] == "Not Available":
                    h["score"] = -1
                else:
                    h["score"] = int(h["score"])
                h["type"] = "hospital"
                hospitals.append(h)
```

```python
cities = []
with open("CitiesInfo.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in tqdm(reader):
        c = {}
#         if row["city_ascii"] == "Atlanta":
#             print(row)
#             print(row["population"])
        if row["population"] != "" and float(row["population"]) > 100000:
            c["state"] = row["state_id"]
            if c["state"] not in ["HI", "AK"]:
                c["pop"] = int(float(row["population"]))
                c["name"] = row["city_ascii"]
                c["lat"] = float(row["lat"])
                c["lng"] = float(row["lng"])
                c["type"] = "city"
                cities.append(c)
print(len(cities))
```

```python
plt.figure(figsize=(20,20))
for h in hospitals[::5]:
    plt.plot(float(h["lng"]), float(h["lat"]), 'bo')
for c in cities:
    plt.plot(float(c["lng"]), float(c["lat"]), 'yo')
plt.show()
```

```python
def badDistance(lat1, lng1, lat2, lng2):
    return ((lat1-lat2)**2 + (lng1-lng2)**2)**(0.5)
```

```python
for h in tqdm(hospitals):
    nearestCity = badDistance(h["lat"], h["lng"], cities[0]["lat"], cities[0]["lng"])
    for c in cities[1:]:
        newBD = badDistance(h["lat"], h["lng"], c["lat"], c["lng"])
        if nearestCity > newBD:
            nearestCity = newBD
    h["nearestCity"] = nearestCity
```

```python
with open('hospitals.json', 'w') as outfile:
    json.dump(hospitals, outfile)
with open('cities.json', 'w') as outfile:
    json.dump(cities, outfile)
```

```python
plt.figure(figsize=(10,10))
scores = []
dists = []
for h in hospitals:
    if h["score"] > -1:
        scores.append(h["score"])
        dists.append(h["nearestCity"])
plt.plot(dists, scores, "ro", alpha=0.05)
plt.show()
```

```python
vrange = np.arange(24, 49+5, 5)
print(vrange)
hrange = np.arange(-125, -65, 5)
print(hrange)
```

```python
chunks = [[[] for v in vrange] for h in hrange]

for h in hospitals:
    cx = math.floor((h["lng"]-hrange[0])/5)
    cy = math.floor((h["lat"]-vrange[0])/5)
    chunks[cx][cy].append(h)

for c in cities:
    cx = math.floor((c["lng"]-hrange[0])/5)
    cy = math.floor((c["lat"]-vrange[0])/5)
    chunks[cx][cy].append(c)

chunkJson = {"myChunks":[]}
for i in range(len(chunks)):
    for j in range(len(chunks[0])):
        chunkJson["myChunks"].append({"name": str(i)+","+str(j), "bodies": chunks[i][j]})

```

```python
with open('chunks.json', 'w') as outfile:
    json.dump(chunkJson, outfile)
```
